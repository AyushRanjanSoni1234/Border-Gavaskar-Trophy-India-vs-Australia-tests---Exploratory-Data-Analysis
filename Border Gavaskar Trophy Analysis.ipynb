```python
Runs=[]
Wics=[]
run_wic=df['team total'].str.split("/").tolist()
print(run_wic)
```

```python
for runs,wics in run_wic:
    Runs.append(runs)
    Wics.append(wics)
```

```python
print(Runs)
```

```python
print(Wics)
```

```python
df["RUNS"]=Runs
df["WICKETS"]=Wics
df["RUNS"]=df["RUNS"].astype('int32')
df["WICKETS"]=df["WICKETS"].astype('int32')
```
